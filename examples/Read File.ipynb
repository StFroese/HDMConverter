# Regenerate Spectrum from File

```python
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
from HDMSpectra import HDMSpectra
```

```python
hdm_spectra = Table.read('../test.dat', format='ascii', delimiter=' ', guess=False)
```

```python
hdm_spectra
```

```python
test_spectrum = hdm_spectra['b'][hdm_spectra['mDM'] == 1000]
log10x = np.unique(hdm_spectra['Log[10,x]'])
x = 10**log10x

true_spec = HDMSpectra.spec('gamma', X='b', xvals=x, mDM=1000, data='../HDMSpectra/data/HDMSpectra.hdf5', annihilation=True)
```

```python
plt.plot(x, test_spectrum)
plt.plot(x, true_spec, linestyle='dashed')
plt.xscale('log')
```
